```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import regex
```

```python
df_amazon_products = pd.read_csv('datasets/products.csv')
#df_amazon_sales = pd.read_csv('datasets/Amazon_Sales_Dataset.csv')
df_amazon_beauty = pd.read_csv('datasets/Amazon_Beauty_Recommendation.csv')
```

```python
df_amazon_products = df_amazon_products.set_index('asin')
df_amazon_products.head()
```

```python
#df_amazon_beauty = df_amazon_beauty.set_index('asin')
df_amazon_beauty.head()
```

```python
text = 'https://www.amazon.in/Dettol-Body-Wash-shower-Refresh/dp/B07XTSQT4R/ref=sr_1_37?crid=3NKGP4OZB9248&keywords=body+wash&pf_rd_i=1374276031&pf_rd_m=A1VBAL9TL5WCBF&pf_rd_p=218f8944-f0f6-41af-ad51-958d2a04fb2f&pf_rd_r=Q92BV6MWDP2BAZXH1E9P&pf_rd_s=merchandised-search-4&qid=1680067965&s=beauty&sprefix=body+%2Cbeauty%2C193&sr=1-37'
r = regex.Regex(r'https://www\.amazon\.in/(.*)/dp/')
def extract_name(text):
    matchs = r.findall(text)
    if len(matchs) > 0:
        return matchs[0].replace('-', ' ')
    else:
        return np.NAN
extract_name(text)
```

```python
df_amazon_beauty_mod = df_amazon_beauty.groupby(by= 'ProductId').min()[['URL']]
df_amazon_beauty_mod['name'] = df_amazon_beauty_mod['URL'].apply(extract_name)
df_amazon_beauty_mod.index.set_names('asin', inplace=True)
df_amazon_beauty_mod.rename(columns={'name':'title'}, inplace=True)
df_amazon_beauty_mod.head()
```

```python
A = set(df_amazon_products.index)
B = set(df_amazon_beauty_mod.index)
C = B.difference(A)
len(A), len(B), len(C)
```

```python
df_all_products = pd.concat([df_amazon_products, df_amazon_beauty_mod[['title']].loc[list(C)]])
df_all_products.head()
```

```python
print(f'Total de datos: {df_all_products.shape[0]}\n')
print(f'Por ciento de datos faltantes:\n{np.round(100 - df_all_products.count() / df_all_products.shape[0] * 100, 2)}')
```

```python
df_all_products.to_csv('datasets/all_products.csv')
```

```python
df_amazon_beauty.groupby(by='UserId')
```

```python
df_sales = df_amazon_beauty.sort_values(by=['UserId', 'Timestamp'], ascending=[True,True])#.set_index(['UserId', 'Timestamp'])
df_sales.head()
```

```python
df_merge = df_sales[['UserId','Timestamp','ProductId']].merge(df_all_products[['title']], left_on='ProductId', right_on='asin')
df_merge.head()
```

```python
view = df_merge.groupby(by='UserId').count()
view = view[(view['Timestamp'] == view['title']) & (view['Timestamp'] > 1)]#.reset_index()
df_sales_merge = df_sales[df_sales['UserId'].isin(view.index)][['UserId','Timestamp','ProductId']]
df_sales_merge.head()
```

```python
df_sequences = pd.DataFrame(df_sales_merge.groupby(by='UserId').apply(lambda x: list(x['ProductId'])), columns=['Sequences'])
df_sequences.head()
```
